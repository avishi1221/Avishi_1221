```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('All_Pokemon_csv.csv')
```

```python

dataset.isnull().sum()
```

```python
data = dataset.copy()
data["Type 2"] = np.where(data["Type 2"].isnull(), 1, 0)
data.groupby("Type 2")['BST'].median().plot.bar()
plt.title("Type 2 median")
plt.show()

data.groupby("Type 2")['BST'].mean().plot.hist()
plt.title("Type 2 mean")
plt.show()
```

```python
dataset.drop(["Type 2","Name"],axis=1,inplace =True)
```

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
print('Number of numerical variables: ', len(numerical_features))
dataset[numerical_features].head()
```

```python
discrete_feature=[]
for feature in numerical_features:
    if len(dataset[feature].unique())<15:
        discrete_feature.append(feature)
print(f"Discrete Variables Count: {len(discrete_feature)}")
discrete_feature_df = pd.DataFrame(discrete_feature, columns=['discrete_features'])
discrete_feature_df
```

```python
for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['BST'].median().plot.bar(color ="purple")
    plt.xlabel(feature)
    plt.ylabel('BST')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature=[]
for feature in numerical_features:
    if feature not in discrete_feature:
        continuous_feature.append(feature)
print(f"Continous Variables Count: {len(continuous_feature)}")
continuous_feature_df= pd.DataFrame(continuous_feature,columns = ["Continuous Features"])
continuous_feature_df
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=20)
    plt.xlabel(feature)
    plt.title(feature)
    plt.show()
```

```python
dataset['BST']=np.log(dataset['BST'])

for feature in continuous_feature:
    data=dataset.copy()
    if 0 in dataset[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['BST'])
        plt.xlabel(feature)
        plt.ylabel('BST')
        plt.title(feature)
        plt.show()
```

```python
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
categorical_features=[]
for feature in dataset.columns:
    if data[feature].dtypes=='O':
        categorical_features.append(feature)
categorical_features_df = pd.DataFrame(categorical_features,columns=["categorical features"])
categorical_features_df
```

```python
categorical_features_df["categories"] = [len(dataset[feature].unique()) for feature in categorical_features]
categorical_features_df
```
