```python
import fitz  # PyMuPDF
import spacy
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load spaCy NLP model
nlp = spacy.load("en_core_web_sm")

# Extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text") + "\n"
    return text.strip()

# Extract Named Entities
def extract_named_entities(text):
    doc = nlp(text)
    entities = [ent.text.lower() for ent in doc.ents]
    return entities

# Extract skills using a simple list of skills (can be expanded)
def extract_skills(text):
    skills = ["python", "machine learning", "tensorflow", "keras", "pytorch", "data science", "nlp", "artificial intelligence"]
    extracted_skills = [skill for skill in skills if skill in text.lower()]
    return extracted_skills

# Job Description for AI
job_description = """
We are looking for a passionate AI engineer with experience in Python, Machine Learning, and Deep Learning frameworks such as TensorFlow, Keras, and PyTorch.
A candidate should have a strong background in algorithms and data structures, NLP techniques, and artificial intelligence. 
Experience with large-scale datasets and model optimization is a plus.
"""

# Define the path to your resume in Kaggle's input folder
input_dir = "/kaggle/input/resume-analyzer"
resume_files = [f for f in os.listdir(input_dir) if f.endswith(".pdf")]

if resume_files:
    pdf_path = os.path.join(input_dir, resume_files[0])  # Use the first PDF found
    resume_text = extract_text_from_pdf(pdf_path)
    
    # Extract named entities and skills from the resume
    resume_entities = extract_named_entities(resume_text)
    resume_skills = extract_skills(resume_text)
    
    # Combine extracted information for comparison
    resume_combined = " ".join(resume_entities + resume_skills)
    
    # Process job description and resume combined text for TF-IDF comparison
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([job_description, resume_combined])

    # Calculate cosine similarity between job description and resume
    similarity_score = cosine_similarity(vectors[0], vectors[1])[0][0]

    # Print results
    print(f"Extracted Resume Skills: {resume_skills}")
    print(f"Resume Entities: {resume_entities}")
    print(f"Job Match Score: {similarity_score:.2f}")
    
    # Estimate chance of selection based on the score
    if similarity_score > 0.8:
        chance_of_selection = "High"
    elif similarity_score > 0.6:
        chance_of_selection = "Medium"
    else:
        chance_of_selection = "Low"
    
    print(f"Chance of Selection: {chance_of_selection}")
else:
    print("No PDF files found in the directory!")
```

```python
def calculate_similarity_score(resume_text, job_description):
    # For example, using cosine similarity between job description and resume text
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.metrics.pairwise import cosine_similarity
    
    # Combine job description and resume text
    documents = [job_description, resume_text]
    
    # Convert texts to TF-IDF features
    vectorizer = TfidfVectorizer().fit_transform(documents)
    
    # Compute cosine similarity between job description and resume
    cosine_sim = cosine_similarity(vectorizer[0], vectorizer[1])
    return cosine_sim[0][0]

def extract_missing_skills(resume_text, job_description):
    # Define a list of required skills based on job description
    job_skills = ['python', 'machine learning', 'tensorflow', 'keras', 'data science', 'artificial intelligence', 'nlp']
    resume_skills = [skill for skill in job_skills if skill.lower() in resume_text.lower()]
    
    # Identify missing skills
    missing_skills = [skill for skill in job_skills if skill not in resume_skills]
    return missing_skills

def generate_feedback(chances_score, missing_skills, experience_years, job_required_experience):
    feedback = ""
    
    # Provide feedback based on similarity score
    if chances_score < 0.5:
        feedback += "Your resume seems to have a low match with the job description. "
    
    # Provide feedback based on missing skills
    if missing_skills:
        feedback += f"You are missing the following skills: {', '.join(missing_skills)}. "
    
    # Provide feedback based on experience mismatch
    if experience_years < job_required_experience:
        feedback += f"Your experience is below the required {job_required_experience} years. Consider gaining more experience in relevant areas. "
    
    # Suggest areas of improvement
    if not feedback:
        feedback = "You seem to be a good fit for the position! Keep up the great work!"
    
    return feedback

def analyze_resume(resume_text, job_description, job_required_experience):
    # Calculate similarity score
    similarity_score = calculate_similarity_score(resume_text, job_description)
    
    # Extract missing skills from the resume
    missing_skills = extract_missing_skills(resume_text, job_description)
    
    # Extract years of experience (for simplicity, we assume experience is given in text like '5 years')
    import re
    experience_years = 0
    experience_match = re.search(r'(\d+)\s*years?', resume_text.lower())
    if experience_match:
        experience_years = int(experience_match.group(1))
    
    # Generate feedback based on analysis
    feedback = generate_feedback(similarity_score, missing_skills, experience_years, job_required_experience)
    
    return feedback

# Example usage
resume_text = "John Doe has a Master's in Computer Science and 3 years of experience in data science, machine learning, and Python."
job_description = "We are looking for someone with experience in Python, machine learning, and at least 5 years of experience in data science."
job_required_experience = 5  # Required years of experience for the job

feedback = analyze_resume(resume_text, job_description, job_required_experience)
print(feedback)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load the pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

def calculate_similarity_score(resume_text, job_description):
    # Encode the resume and job description
    sentences = [resume_text, job_description]
    embeddings = model.encode(sentences)
    
    # Compute cosine similarity
    similarity = cosine_similarity([embeddings[0]], [embeddings[1]])
    return similarity[0][0]

# Example usage
similarity_score = calculate_similarity_score(resume_text, job_description)
print(f"Improved Similarity Score: {similarity_score}")
```

```python
def provide_feedback(score):
    if score >= 0.8:
        return "Your resume is a strong match for the job description! You're a good fit."
    elif score >= 0.5:
        return "Your resume matches the job description moderately well. There are a few areas where you could improve, such as detailing your experience with specific tools or platforms."
    else:
        return "Your resume does not match the job description well. Consider improving your experience section or gaining expertise in the required areas like AI/ML frameworks, cloud services, or project experience."

# Example usage:
feedback = provide_feedback(similarity_score)
print(feedback)
```
