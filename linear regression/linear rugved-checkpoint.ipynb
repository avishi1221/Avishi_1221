```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv')
```

```python
data = data.drop(columns=['depth','table'])
```

```python
data = pd.get_dummies(data).astype(int)
data = data.drop(columns=['cut_Fair','color_D','clarity_I1'])
data.head(5)
```

```python
for column in data.columns:
    mini = data[column].min()
    maxi = data[column].max()
    data[column] = (data[column] - mini)/(maxi - mini)       
data.head(3)
```

```python
data_indices = list(range(len(data)))
test_size = int(0.2 * len(data))
data_train = data.iloc[data_indices[test_size:]]
data_test = data.iloc[data_indices[:test_size]]
```

```python
X_train = data_train.drop(columns=['price'])
Y_train = data_train['price']

X_test = data_test.drop(columns=['price'])
Y_test = data_test['price']
```

```python
no_of_iterations = 1000
learning_rate = 0.01
m, n = X_train.shape
w = np.zeros(n)
b=0
for i in range(no_of_iterations):
    Y_prediction = X_train.dot(w) + b
    dw = - ( 2 * (X_train.T).dot(Y_train - Y_prediction)) / m
    db = - 2 * np.sum(Y_train - Y_prediction) / m 
    w = w - learning_rate * dw
    b = b - learning_rate * db
```

```python
result = X_test.dot(w) + b
```

```python
RMSE = 0.0
for i in range(len(result)):
    RMSE += (Y_test.iloc[i] - result.iloc[i]) ** 2 
    RMSE = (RMSE / len(result)) ** 0.5
    VarY =0.0
    mean = Y_test.mean()
    for i in range(len(result)):
        VarY += (Y_test.iloc[i]-mean)**2
    R2 = 1 - ((RMSE**2)/VarY)
R2*100
```
